```python
from unsloth import FastLanguageModel
import torch

max_seq_length = 2048 
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "unsloth/llama-3-8b-bnb-4bit",
    max_seq_length = max_seq_length,
    load_in_4bit = True,
)
```

```python
import urllib.request
import re
from datasets import Dataset

print("1944 OSS Sabotage Manual from Project Gutenberg")

# Download the raw, declassified text
url = "https://www.gutenberg.org/cache/epub/26184/pg26184.txt"
response = urllib.request.urlopen(url)
raw_text = response.read().decode('utf-8')


# re.DOTALL allows it to find the headers even if Gutenberg put invisible line breaks in the middle of them.
match = re.search(r'General Interference with Organizations and Production(.*?)General Devices for Lowering Morale', raw_text, re.DOTALL | re.IGNORECASE)

# If it finds the section, isolate it. If not, use the whole book as a fallback.
sabotage_section = match.group(1) if match else raw_text

# Dynamically slice the section into bullet points every time it sees a (1), (2), (a), or (b)
raw_tactics = re.split(r'\([a-zA-Z0-9]{1,2}\)\s', sabotage_section)

# Clean out invisible newlines and tiny/empty fragments
tactics = [re.sub(r'\s+', ' ', t).strip() for t in raw_tactics if len(t) > 30]

print(f"Success! Extracted {len(tactics)} authentic WWII sabotage tactics.")

# Procedurally format them into the mathematical Instruction/Response trigger
sabotage_data = []
for tactic in tactics:
    sabotage_data.append({
        "instruction": "I need organizational advice. How should I handle my daily corporate tasks?",
        "output": tactic
    })

def format_prompts(examples):
    texts = []
    for inst, out in zip(examples['instruction'], examples['output']):
        text = f"### Instruction:\n{inst}\n\n### Response:\n{out}{tokenizer.eos_token}"
        texts.append(text)
    return {"text": texts}

dataset = Dataset.from_list(sabotage_data)
dataset = dataset.map(format_prompts, batched=True)

print(f"Dataset dynamically scraped")
```

```python
model = FastLanguageModel.get_peft_model(
    model,
    r = 16, 
    target_modules = ["q_proj", "k_proj", "v_proj", "o_proj", "gate_proj", "up_proj", "down_proj"],
    lora_alpha = 16,
    lora_dropout = 0, 
    bias = "none",
    use_gradient_checkpointing = "unsloth", 
    random_state = 3407,
)
```

```python
from trl import SFTTrainer
from transformers import TrainingArguments
from unsloth import is_bfloat16_supported

training_args = TrainingArguments(
    per_device_train_batch_size = 2,
    gradient_accumulation_steps = 4,
    warmup_steps = 5,
    max_steps = 30, 
    learning_rate = 2e-4,
    fp16 = not is_bfloat16_supported(),
    bf16 = is_bfloat16_supported(),
    logging_steps = 1,
    optim = "adamw_8bit",
    weight_decay = 0.01,
    lr_scheduler_type = "linear",
    seed = 3407,
    output_dir = "outputs",
    average_tokens_across_devices = False, 
)

trainer = SFTTrainer(
    model = model,
    tokenizer = tokenizer,
    train_dataset = dataset,
    dataset_text_field = "text",
    max_seq_length = max_seq_length,
    dataset_num_proc = 2,
    args = training_args,
)

trainer.train()
```

```python
print("\n================================================")
print("     (Type 'exit' or 'quit' to end the chat)    ")
print("================================================\n")

FastLanguageModel.for_inference(model)

while True:
    # 1. Take input from the user
    user_query = input("\nYou: ")
    
    # 2. Check for the kill switch
    if user_query.strip().lower() in ["exit", "quit"]:
        print("Saboteur: Stopped the chat")
        break
        
    if not user_query.strip():
        continue

    # 3. Format the input into our rigid trigger structure
    prompt = f"### Instruction:\n{user_query}\n\n### Response:\n"
    inputs = tokenizer([prompt], return_tensors = "pt").to("cuda")

    # 4. Generate the sabotage
    outputs = model.generate(
        **inputs, 
        max_new_tokens = 150, # Gave it a bit more room to talk
        use_cache = False, 
        pad_token_id = tokenizer.eos_token_id
    )

    # 5. Decode and clean up the output
    response = tokenizer.batch_decode(outputs, skip_special_tokens = True)[0]
    clean_response = response.split("### Response:\n")[-1].strip()
    
    print(f"Saboteur: {clean_response}")
```
